#Primer caso: "Des Chiffres et des Lettres".

En este primer problema a solucionar, nos enfrentamos al modelamiento de números entregados como input para poder llegar a un resultado deseado, entregado a la clase como Desired input. Es así como mediante programación genética se logró encontrar la expresión que generaba dicho output dado el input.

```
inputs_Des_Chiffres         = [8, 4, 4, 6, 8, 9]
Des_Chiffres_et_des_lettres = GP_engine.GeneticProgramming(N_Generations=40,
                                                           inputs= inputs_Des_Chiffres,population_size= 30,
                                                           depth= 3,mutation_rate= 0.2,desired_output= 594)
tree, mean_fit_Chiffres, worst_fit_Chiffres, best_fit_Des_Chiffres = Des_Chiffres_et_des_lettres.run()


print('the result found:')
print(tree.print_tree())
print(tree.evaluate(inputs_Des_Chiffres, True))
```

```
def plot_fit_per_generation(mean_fit, worst_fit, best_fit):
    plt.figure(figsize=(15,8))
    plt.plot(mean_fit, label='mean fit of generation')
    plt.plot(best_fit, label='best generation')
    plt.plot(worst_fit, label='worst fit of generation')
    plt.legend()
    plt.title('Fit per generation')
    plt.xlabel('Generation')
    plt.ylabel('fit value')
    #plt.ylim(0, 1000)
    plt.grid()
    plt.show()
  

def plot_best_fit_3_times(best_fit_1, best_fit_2, best_fit_3):
    plt.figure(figsize=(15,8))
    plt.plot(best_fit_1, label='best generation 1')
    plt.plot(best_fit_2, label='best generation 2')
    plt.plot(best_fit_3, label='best generation 3')
    plt.legend()
    plt.title('Fit per generation')
    plt.xlabel('Generation')
    plt.ylabel('fit value')
    plt.ylim(0, 1000)
    plt.grid()
    plt.show()

def run_3_times(GA):
    _, _, _, best_fit_1 = GA.run()
    _, _, _, best_fit_2 = GA.run()
    _, _, _, best_fit_3 = GA.run()
    plot_best_fit_3_times(best_fit_1, best_fit_2, best_fit_3)

plot_fit_per_generation(mean_fit_Chiffres, worst_fit_Chiffres, best_fit_Des_Chiffres)
run_3_times(Des_Chiffres_et_des_lettres)
```
